```python
import pandas as pd

# Load the CSV file with max memory utilisation
py_df = pd.read_csv('data/2020_sharing_data_outside.csv', low_memory=False)

py_df.head()
```

```python
# General Columns

general_columns = ['age',
                    'are.you.datascientist',
                    'is.python.main',
                    'company.size',
                    'country.live',
                    'employment.status',
                    'first.learn.about.main.ide',
                    'how.often.use.main.ide',
                    'is.python.main',
                    'main.purposes'
                    'missing.features.main.ide'
                    'nps.main.ide',
                    'python.version.most',
                    'python.years',
                    'python2.version.most',
                    'python3.version.most',
                    'several.projects',
                    'team.size',
                    'use.python.most',
                    'years.of.coding'
]


def column_multi_name(column_name):
    """Check if column is in the general and return (general, column_name)
        Else, return two-element tuple from str.rsplit
    """
    if column_name in general_columns:
        return ('general', column_name)
    
    else:
        first, rest = column_name.rsplit('.', 1)
        return(first, rest)
    
    
```

```python
# Using multi-index columns

py_df.columns = pd.MultiIndex.from_tuples([column_multi_name(col) for col in py_df.columns])
```

```python
# Sort the columns in alphabetical order

py_df = py_df[sorted(py_df.columns)]

# Preview the dataframe
py_df.head()
```

```python
# Since this is a multi-Index col, lets find most used IDE

py_df[('ide', 'main')].value_counts(normalize=True).sort_values(ascending=False)
```

```python
# Other languages most commonly used by Python Developers
py_df[('other.lang')].count().sort_values(ascending=False)
```

```python
# Countries with the greatest number of survey respondents

py_df['general', 'country.live'].value_counts(normalize=True).sort_values(ascending=False)
```

```python
# Proportion of Developers with level of experience

py_df['general', 'python.years'].value_counts(normalize=True).sort_values(ascending=False)
```

```python
# The country with the most experienced developers

py_df['general'][py_df[('general' , 'python.years')] == '11+ years'].groupby('country.live')['python.years'].count().sort_values(ascending=False)
```

```python
# Which country has the greatest proportion of Python Developers with 11+ years of experienced

country_experience = py_df['general'][['country.live', 'python.years']]

# All developers per country
all_per_country = country_experience['country.live'].value_counts()

# Expert per country
expert_per_country = country_experience.loc[(country_experience['python.years'] == '11+ years')]['country.live'].value_counts()

# Now we divide the expert per country by the all devs in a country
(expert_per_country / all_per_country).sort_values(ascending=False).dropna().head(10)
```

```python
# Stackoverflow Survey

so_df = pd.read_csv('data/so_2021_survey_results.csv', low_memory=False)

so_df.head()
```

```python
# Do Free Lancers earn more than fulltime employees?

so_df.groupby('Employment')['ConvertedCompYearly'].mean().sort_values(ascending=False).dropna().apply(lambda n: f'{n:,.2f}')
```

```python
# Pivot Table of Country names and Ed Level with Salaries

so_df.pivot_table(index='Country', columns='EdLevel', values='ConvertedCompYearly', aggfunc='mean')
```

```python
oecd_df = pd.read_csv('data/oecd_locations.csv', header=None, index_col=1, names=['abbrev', 'Country'])

oecd_df.head()
```

```python
# 

oecd_df.join(so_df.set_index('Country').pivot_table(index='Country', columns='EdLevel', values='ConvertedCompYearly', aggfunc='mean'))
```

```python
so_df = so_df.loc[~so_df['LanguageHaveWorkedWith'].isna()]

so_df = so_df[so_df['LanguageHaveWorkedWith'].str.contains('Python')]

so_df.shape
```

```python
# Let all less than 1 year = 0
so_df.loc[(so_df['YearsCode'] == 'Less than 1 year'), 'YearsCode'] = 0

# Let all More than 50 years be 51
so_df.loc[(so_df['YearsCode'] == 'More than 50 years'), 'YearsCode'] = 51

# Convert col YearsCode to int
so_df['YearsCode'] = so_df['YearsCode'].dropna().astype(int)

so_df['experience'] = pd.cut(so_df['YearsCode'], 
                             bins=[-1, 1, 2, 5, 10, 100], 
                             labels=['Less than 1 year', '1-2 years', '3-5 years', '6-10 years', '11+ years'])


```
